```python
data_dir = "/Users/tangyujin/Desktop/09-半监督/03_data/ISIC2019 384x384 jpeg/train/"
TRAIN_dir = "/Users/tangyujin/Desktop/09-半监督/03_data/ISIC2019 384x384 jpeg/processed/train/"
TEST_dir = "/Users/tangyujin/Desktop/09-半监督/03_data/ISIC2019 384x384 jpeg/processed/test/"
VAL_dir = "/Users/tangyujin/Desktop/09-半监督/03_data/ISIC2019 384x384 jpeg/processed/val/"
```

```python
import pandas as pd
import numpy as np
import cv2
from skimage import color
from skimage import io
from scipy import misc
import imageio
import time
from tqdm import tqdm_notebook
from glob import glob
import os
```

```python
def grey_world(nimg):
    nimg = nimg.transpose(2, 0, 1).astype(np.uint32)
    mu_g = np.average(nimg[1])
    nimg[0] = np.minimum(nimg[0]*(mu_g/np.average(nimg[0])),255)
    nimg[2] = np.minimum(nimg[2]*(mu_g/np.average(nimg[2])),255)
    return nimg.transpose(1, 2, 0).astype(np.uint8)

def process_train(name): # performs same steps as process_image but saves img in TRAIN
    img = io.imread(data_dir+name+".jpg")
    img = grey_world(img)
    img = cv2.resize(img, dsize=(224, 224))#, interpolation=cv2.INTER_CUBIC)  
    imageio.imsave(TRAIN_dir+ name+".jpg", img)

def process_test(name): # takes filename of the lesion sample
    img = io.imread(data_dir+name+".jpg")
    img = grey_world(img)
    img = cv2.resize(img, dsize=(224, 224))#, interpolation=cv2.INTER_CUBIC)
    imageio.imsave(TEST_dir+ name+".jpg", img)
def process_val(name): # performs same steps as process_image but saves img in TRAIN
    img = io.imread(data_dir+name+".jpg")
    img = grey_world(img)
    img = cv2.resize(img, dsize=(224, 224))#, interpolation=cv2.INTER_CUBIC)  
    imageio.imsave(VAL_dir+ name+".jpg", img)
```

```python
import time
from tqdm import tqdm_notebook
start = time.time()
for lesion in tqdm_notebook(TRAIN.index):
    process_train(lesion)
    
end = time.time()
print("Run time for "+ str(len(TRAIN.index)) + " images: " + str((end - start)/60)+" minutes")    
```

```python
import time
from tqdm import tqdm_notebook
start = time.time()
for lesion in tqdm_notebook(TEST.index):
    process_test(lesion)
    
end = time.time()
print("Run time for "+ str(len(TEST.index)) + " images: " + str((end - start)/60)+" minutes")    
```

```python
import time
from tqdm import tqdm_notebook
start = time.time()
for lesion in tqdm_notebook(VAL.index):
    process_val(lesion)
    
end = time.time()
print("Run time for "+ str(len(VAL.index)) + " images: " + str((end - start)/60)+" minutes")    
```
